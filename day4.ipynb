```python
import re
import datetime
import pandas as pd

data = []
with open("day4.input") as f:
    for l in f:
        data.append(l.strip())
    #data = f.read()


f = []
for d in data:
    g = re.match(r"\[(\d+)-(\d+)-(\d+) (\d+):(\d+)\] (.*)$", d)

    #print(g.group(0), g.group(1) , g.group(2), g.group(3), g.group(4), g.group(5), g.group(6))
    #print(datetime.datetime(int(g.group(1)), int(g.group(2)), int(g.group(3)), int(g.group(4)), int(g.group(5))))
    f.append((datetime.datetime(int(g.group(1)), int(g.group(2)), int(g.group(3)), int(g.group(4)), int(g.group(5))),\
              g.group(6), (int(g.group(2)), int(g.group(3))), int(g.group(5))) )

#print(pd.DataFrame(f))

#df = pd.DataFrame(f, columns=["dates", "value"]) 
#print(df)

#print(df.head(5))
#df = df.sort_values("dates")
#print(df)

#for i, r in df.iterrows():
#    print(r)
```

```python
df = pd.DataFrame(f, columns=["dates", "value", "month/day", "time"]) 
#print(df)

#print(df.head(5))
df = df.sort_values("dates")
#print(df)
#display(df.head())

cur_id = 0
for i, r in df.iterrows():
    mat = re.match(r"Guard #(\d+)", r["value"])
    if mat:
        #print(mat.group(1))
        cur_id = mat.group(1)
        df.loc[i, "id"] = 0
        #print(cur_id)
    else:
        df.loc[i, "id"] = cur_id
    #print(r)

#display(df['dates'].dt.time)

#display(df[df["id"] != 0])#.groupby(["month/day", "id"]).agg("sum"))
```

```python

sleep = False
sleep_time = 0
df2 = df[df["id"] != 0]
for i, r in df2.iterrows():
    if r['value'] == "falls asleep":
        sleep = True
        sleep_time = r['time'] 
        df2.loc[i, "sleep_time"] = r['time']
    if sleep == True and r['value'] == "wakes up":
        sleep = False
        df2.loc[i, "sleep_time"] = str(list(range(sleep_time, r['time'])))
        sleep_time = -1
        #df2.loc[i, "wake_time"] = r['time'] #(sleep_time, r['time']) 

display(df2[df2["value"] != "falls asleep"])
```
